## 三、对数据进行处理

### 3.1利用纯Python代码对时区进行计数

　　我们想要知道数据集中最常出现的是哪个时区（即tz），此处使用列表解析式从数据中取出时区tz对应的数据。

```python
time_zones = [rec['tz'] for rec in records]
```

　　以上返回KeyError错误，说明并非所有记录中都包含时区字段。故需要对其进行条件判断。

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
len(time_zones) # len(records)为3560，即有120条记录中不包含时区字段
```

```python
time_zones[:10] # 提取前10条记录中的时区字段（其中部分时区字段为空值）
```

　　对时区进行计数，可用标准库，但比较难；可用pandas库，比较简单；还可以自己编写循环进行计数。

#### 1、编写函数遍历时区将计数值保存在字典中

```python
def get_counts(seq):
    counts = {}  # 建立计数空字典
    for x in seq: # 遍历指定序列
        if x in counts: 
            counts[x] += 1 # 若序列中某个值在计数字典中再次出现则对应的计数值加1
        else:
            counts[x] = 1 # 若未再次出现则计数值为1
    return counts # 返回计数字典

counts = get_counts(time_zones)
counts[''] # 通过键查看对应时区出现次数
```

#### 2、使用标准库进行计数

```python
from collections import defaultdict

def get_counts(seq):
    counts = defaultdict(int) # 所有值均会被初始化为零
    for x in seq:
        counts[x] += 1
    return counts

get_counts(time_zones)
```

#### 3、根据以上对时区的统计，获取计数值前10的时区

```python
def top_counts(count_dict, n = 10):
    
    # 根据“.items()”获取字典的键值，并将其转换为值-键形式
    value_key_pairs = [(count, tz) for tz, count in count_dict.items()]
    
    # 对其进行排序
    value_key_pairs.sort()
    
    # 返回索引由-10到最后的值
    return value_key_pairs[-n:]

top_counts(counts)
```

```python
# 使用标准库简化该任务

from collections import Counter

counts = Counter(time_zones) # 统计各个时区出现的数值
counts.most_common(10) # 查找出现最多次数的时区
```

### 3.2 利用pandas库对时区进行计数

```python
import pandas as pd 
import numpy as np
```

```python
frame = pd.DataFrame(records)
frame['tz'][:10]
```

　　**".value_counts"**函数可用于统计每个时区出现的次数（类似于上述标准库中的**"Counter"函数**）,该统计结果按照多到少顺序排列

```python
tz_counts = frame['tz'].value_counts()
tz_counts[:10]
```

### 4、可视化

　　由于部分时区字段为空值，故进行可视化之前需要进行预处理，填补缺失值，**pandas库中fillna函数**可用于替换缺失值NaN。

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
clean_tz = frame['tz'].fillna('Missing')
clean_tz[clean_tz == ''] = 'Unknown'
tz_counts = clean_tz.value_counts()
tz_counts[:10]
```

```python
# kind表示图形类别，“barh”表示横向的柱状图，rot表示各个名称的旋转角度，“0”即表示旋转零度
tz_counts[:10].plot(kind = 'barh', rot = 0)
plt.show()
```

### 5、对字符串进行处理

#### 1、将字符串中第一节（与浏览器对应）分别出来

```python
# 将a中字符串按照空格进行分割，“dropna”用与去除缺失值,只能对当行进行处理
results = pd.Series([x.split()[0] for x in frame.a.dropna()])
results.head()
```

```python
results.value_counts().head()
```

#### 2、按Windows用户和非Windows用户对时区进行分解

　　只需要“a”中字符串含有“Windows”即认为该用户为Windows用户，首先需移除缺失值。

```python
# isnull、notnull返回布尔值，若为缺失值则返回True
clean_a = frame[frame.a.notnull()]
```

　　**np.where(condition, [x, y])**：根据条件判断，若符合条件则返回x，反之返回y。

```python
operating_system = np.where(clean_a['a'].str.contains('Windows'), 'Windows', 'Not Windows')
operating_system[:5]
```

```python
# tz按照operating_system进行统计
by_tz_operating_system = clean_a.groupby(['tz', operating_system])

# 获取不同操作系统对应tz总数，利用unstack重塑，并将缺失值以0填充
agg_counts = by_tz_operating_system.size().unstack().fillna(0)
agg_counts.head()
```

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot(kind = 'barh', stacked = True)
plt.show()
```

```python
normed_subset = count_subset.div(count_subset.sum(1), axis = 0)
normed_subset.plot(kind = 'barh', stacked = True)
plt.show()
```
